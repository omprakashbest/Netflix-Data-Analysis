```python
# had to replace the 'Not Available' and 'unknown' with NaN

df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
```

```python
# rename columns names... make them title_and replace spaces with underscores

df.columns = df.columns.str.title().str.replace(' ', '_')
```

```python
df.columns
```

```python
# derive the new columns discount, List_price and Discount_percent 

df['Discount'] = df['List_Price'] * df['Discount_Percent'] * 0.01 
```

```python
# derive the new columns sale_price and profit

df['Sale_Price'] = df['List_Price'] - df['Discount']

df['Profit'] = df['Sale_Price'] - df['Cost_Price']
```

```python
# dtypes pf dataset : 

df.info()
```

```python
# Convert the date from object dtype to datetime dtype.

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%Y-%M-%D')
```

```python
df.dtypes
```

```python
# drops columns -> cost price and list price and discount percent

df.drop(columns=['Cost_Price', 'List_Price', 'Discount_Percent'], inplace=True)
```
